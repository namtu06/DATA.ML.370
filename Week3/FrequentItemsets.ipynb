## Task 3.1b

Algorithm constructed with support from Copilot

1. Generate baskets and get their supports
2. Generate L1 (frequent 1-itemsets)
3. Generate C2 (candidate doubles)
4. Compute supports for C2
5. Generate L2
6. Generate C3
7. Compute supports for C3
8. Generate C4
9. Code stops because no new frequent itemsets are found

```python
from itertools import combinations

#Generate baskets
def generate_baskets(n=100):
    baskets = []
    for b in range(1, n + 1):
        basket = set()
        for i in range(1, n + 1):
            if b % i == 0:
                basket.add(i)
        baskets.append(basket)
    return baskets

#Count support
def get_support(itemset, baskets):
    count = 0
    for basket in baskets:
        if itemset.issubset(basket):
            count += 1
    return count

#Apriori algorithm
def apriori(baskets, min_support):
    # Generate 1-itemsets
    items = set()
    for basket in baskets:
        items |= basket

    freq_itemsets = []
    L1 = []
    
    for item in items:
        support = get_support({item}, baskets)
        if support >= min_support:
            L1.append(frozenset([item]))
    
    current_L = L1
    k = 2


    #Generate C2 then L2 then C3 then L3 so on and son
    while current_L:
        freq_itemsets.extend(current_L)

        # Generate candidates
        candidates = set()
        for a, b in combinations(current_L, 2):
            union = a | b
            if len(union) == k:
                candidates.add(union)

        # Prune step
        pruned_candidates = set()
        for candidate in candidates:
            all_subsets_frequent = True
            for subset in combinations(candidate, k - 1):
                if frozenset(subset) not in current_L:
                    all_subsets_frequent = False
                    break
            if all_subsets_frequent:
                pruned_candidates.add(candidate)

        # Count support
        next_L = []
        for candidate in pruned_candidates:
            support = get_support(candidate, baskets)
            if support >= min_support:
                next_L.append(candidate)

        current_L = next_L
        k += 1

    return freq_itemsets

# Run everything
baskets = generate_baskets(100)
min_support = 5

frequent_itemsets = apriori(baskets, min_support)

# Print results
print(", ".join("{" + ", ".join(map(str, sorted(itemset))) + "}" for itemset in frequent_itemsets))
```

## Task 3.2.1

FP-Growth (Frequent-Pattern-Growth) Algorithm also mines for frequent itemsets but unlike Aprior, it is not as computationally costly by compressing data in to a FP-tree and extracting patterns directly from it only with 2 database scans

1. Data compression: Compresses data into a FP-tree that stores information about itemsets and their frequencies without needing to generate candidate sets like Apriori

2. Tree Mining: Examines tree to find patterns that appear  based on minimum threshold, the algorithm does this via breaking it down into smaller "conditional" trees for each item

3. Generate patterns: Generate frequent itemsets and rules describing relationships

## Task 3.2.2

### These files contain 1,000,209 anonymous ratings of approximately 3,900 movies made by 6,040 MovieLens users who joined MovieLens in 2000.

### At only(?) 35% threshold for minimum support, the list of frequent itemsets only has 2 members left. It can be seen from the few test thresholds that for a large dataset, a few percentage points can significantly alter the amount of frequent itemsets. Additionally, computing these sets with FPGrowth took a significantly shorter amount of time than compared to Aprior in the last task.

```python
import os
import pyspark
from pyspark.sql import *
sc = pyspark.SparkContext("local", "FrequentItemsets")
os.system('wget http://files.grouplens.org/datasets/movielens/ml-1m.zip; unzip ml-1m.zip')
```

```python
from pyspark.mllib.fpm import FPGrowth
lines = sc.textFile("ml-1m/movies.dat")
movienames = lines.map(lambda l: l.split("::")). \
    map(lambda p: ( int(p[0]), p[1] ) ).collect()
#convert to dictionary for printing lookup
movienames = dict( movienames )
#read ratings file as lines of text, assuming no errors on data
lines = sc.textFile("ml-1m/ratings.dat")
#group user ratings into (user, list of rated movie ids)
usermovies = lines.map(lambda l: l.split("::")). \
    map(lambda p: ( int(p[0]), int(p[1]) ) ).groupByKey()
#take only list part
usermovies = usermovies.map(lambda p: list(p[1]))
usercount = usermovies.count()
model = FPGrowth.train(usermovies, minSupport=0.2, numPartitions=10)
result = model.freqItemsets()
```

```python
for x in result.collect():
    keys = x[0]
    if len(keys)>4:
        fpairs = []
        for y in keys:
            fpairs.append(movienames[y])
        print(fpairs)
        print("Found in " + str(x[1]) + " user of " + str(usercount) + "\n")
```
